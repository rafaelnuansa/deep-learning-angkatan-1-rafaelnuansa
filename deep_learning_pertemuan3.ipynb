# Fungsi Aktivasi

Fungsi aktivasi merupakan komponen penting dalam jaringan saraf tiruan. Jadi fungsi ini menentukan apakah neuron harus diaktifkan (menghasilkan nilai) atau tidak, berdasarkan bobot masukannya.

## Sigmoid Activation Function

$$
  f(x) = \frac{1}{1+e^{-x}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def rumus_sigmoid(x):
    return 1/ (1+np.exp(-x))

x = np.linspace(-5,5,100)
y = rumus_sigmoid(x)

plt.plot(x,y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Sigmoid Activation Function')
plt.grid(True) #ada kotak kotaknya
plt.show()

#Semakin jauh range-nya, semakin tegak fungsi aktivasinya (yang menuju ke 0 & 1)
```
